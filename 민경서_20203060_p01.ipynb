```python
#Task3-2
metanew = meta[['location', 'date', 'total_deaths_per_million', 'total_cases_per_million', 'people_vaccinated_per_hundred', 'human_development_index']]
metanew = metanew.groupby('location').max()
metanew 
```

```python
#Task3-3
import matplotlib.pyplot as plt

total = (metanew.total_deaths_per_million / metanew.total_cases_per_million)
total.dropna(how='any')
plt.scatter(metanew.people_vaccinated_per_hundred, total)

plt.xlabel("people_vaccinated_per_hundred")
plt.ylabel("deaths_per_cases")
plt.show()
```

```python
total = (metanew.total_deaths_per_million / metanew.total_cases_per_million)
total
```

```python
#Task3-4 상관계수 구하기
import sys
metanew = metanew.dropna(how='any')
total = total.dropna(how='any')
mean_x = metanew.people_vaccinated_per_hundred.mean()
#print(mean_x)
mean_y = total.mean()
#print(mean_y)

std_x = (sum((x - mean_x) ** 2 for x in metanew.people_vaccinated_per_hundred)/len(metanew.people_vaccinated_per_hundred)) ** 0.5 
#print(std_x)
std_y = (sum((y - mean_y) ** 2 for y in total)/len(total)) ** 0.5 
#print(std_y)
cov = sum((x - mean_x) * (y - mean_y) for x, y in zip(metanew.people_vaccinated_per_hundred, total))/len(metanew.people_vaccinated_per_hundred)
#print(cov)
corr = cov/std_x/std_y
corr
```

```python
#Task4-1 상관계수 구하기
import sys
metanew = metanew.dropna(how='any')
#total = total.dropna(how='any')
mean_x = metanew.human_development_index.mean()
#print(mean_x)
mean_y = metanew.people_vaccinated_per_hundred.mean()
#print(mean_y)

std_x = (sum((x - mean_x) ** 2 for x in metanew.human_development_index)/len(metanew.human_development_index)) ** 0.5 
#print(std_x)
std_y = (sum((y - mean_y) ** 2 for y in metanew.people_vaccinated_per_hundred)/len(metanew.people_vaccinated_per_hundred)) ** 0.5 
#print(std_y)
cov = sum((x - mean_x) * (y - mean_y) for x, y in zip(metanew.human_development_index, metanew.people_vaccinated_per_hundred))/len(metanew.human_development_index)
#print(cov)
corr = cov/std_x/std_y
corr
```

```python
#Task4-2 linear regression 
import torch

x_train = torch.FloatTensor([metanew.human_development_index])
y_train = torch.FloatTensor([metanew.people_vaccinated_per_hundred])
W = torch.zeros([1])
b = torch.zeros([1])
lr = 0.01
for epoch in range(3001):
    W.requires_grad_(True)
    b.requires_grad_(True)

    hypothesis = torch.mm(x_train, W) + b
    cost = torch.mean((hypothesis - y_train) ** 2)
    cost.backward()
    with torch.no_grad() as grd:
        W = W - lr * W.grad
        b = b - lr * b.grad

    if epoch % 100 == 0:
        print( 'epoch: {}, cost: {:.6f}, W: {}, b: {}'.format(epoch, cost.item(), W.squeeze(), b))

```

```python

from sklearn.linear_model import LinearRegression
x = metanew.human_development_index
y = metanew.people_vaccinated_per_hundred
lr = LinearRegression() 
#lr.fit(x, y)
lr.fit(x.values.reshape(-1,1), y)

#학습 결과
print(lr.coef_, lr.intercept_)
```
